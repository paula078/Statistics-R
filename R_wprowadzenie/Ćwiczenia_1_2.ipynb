## Podstawowa struktura danych - “Data frame”

**Data frame** to tabela, a więc obiekt podobny do dwuwymiarowej macierzy, ale mająca pewne szczególne cechy, kóre sprawiają że stanowi dobrze zorganizowaną bazę danych:

- Ma nagłówek kolumn, a więc każda kolumna ma swoją unikatową nazwę. Jeśli nazwy nie są wskazane w danych źródłowych, R da im nazwy automatycznie - V1, V2, ….
- Tabela jest zawsze kompletna - każda kolumna ma tyle samo wierszy i odwrotnie.
- Każda ‘komórka’ zawiera jakąś pojedynczą wartość. Jeśli dla jakiejś komórki brak danych, zostanie w niej wpisane NA (“Not Available”).
- Każda kolumna jest w zasadzie wektorem (tyle, że z towarzyszącą mu nazwą) i może zawierać tylko jeden, określony typ danych. Różne kolumny mogą zawierać dane różnego typu.
- Jeden wiersz odpowiada jednemu ‘rekordowi’ w bazie danych, a więc zawiera komplet informacji dla jakiegoś obiektu bądź zdarzenia.

## Import danych do ‘Data frame’ i podgląd danych
Dane importowane są zwykle w postaci plików tekstowych, typowo mających rozszerzenia:

- .csv - comma-separated values - linie tekstu odpowiadają wierszom tabeli, a kolumny są rozdzielone przecinkami.
- .tsv - tab-separated values - kolumny rozdzielane tabulatorami.
- .txt - plik tekstowy o nieokreślonej strukturze

**UWAGA** Niestety, w polskim standardzie zapisu wartośći liczbowych, przecinków używa się do oznaczenia miejsc dziesiętnych, a kolejne wartości rozdzielane są zwykle średnikami. Dlatego nie można być pewnym, że plik .csv, wygenerowany np. przy pomocy Excela, ma wartości rzeczywiście rozdzielane przecinkami.

### Importowanie danych z pliku CO2.csv i zapisanie w zmiennej typu data frame

- "CO2.csv" to nazwa pliku, który ma zostać wczytany.

- header=TRUE oznacza, że pierwszy wiersz pliku zawiera nagłówki kolumn.

- sep="," oznacza, że przecinek (,) jest używany jako separator pomiędzy wartościami w pliku.

- quote="\"" oznacza, że cudzysłów (") jest używany jako znak zamykający dla wartości, które zawierają przecinki lub nowe linie.

- dec="." oznacza, że kropka (.) jest używana jako separator dziesiętny w pliku.

- fill=TRUE oznacza, że jeśli wiersze w pliku mają różną ilość wartości, brakujące wartości zostaną uzupełnione wartością pustą.

- comment.char="" oznacza, że komentarze nie są używane w pliku (jeśli byłyby, linie zaczynające się od tego znaku byłyby ignorowane).

```r
CO2_data <- read.table("CO2.csv", header = TRUE, sep=",", quote="\"", dec=".", fill= TRUE, comment.char="")
CO2_data
```

```r
head(CO2_data, n=5)
tail(CO2_data, n=2)

View(CO2_data)
```

## Typy danych
Dana kolumna może zawierać tylko jeden, określony typ danych. Różne kolumny mogą zawierać dane różnego typu. Typ danych dla danej kolumny jest definiowany przez ich klasę (class). Najczęściej używanymi klasami są:

- character - czyli łańcuchy znaków (liczby otoczone cudzysłowami traktowane są też jako ‘strings’)

- factor - ‘strings’ którym przypisane są uporządkowane ‘poziomy’ (kolejne liczby naturalne) i do których dowiązana jest informacja o liczbie wystąpień danego ‘poziomu czynnika’ w zborze danych. ‘Czynniki’ wykorzystywane są powszechnie w analizach staystycznych, w których reprezentują wartości w skali nominalnej (sprawdź też informacje o typach danych w analizach statystycznych. Żeby ‘strings’ zostały potraktowane jako ’factors" należy użyć w funkcji read.table() argumentu stringAsFactors = TRUE (dla wersji R >= 4.0.0).

- numeric - liczby rzeczywiste

- integer - liczby całkowite

- logical - wartosći logiczne: TRUE (T) lub FALSE (T)

Klasa danych może być zmieniona funkcjami z rodziny as…, np. as.character(), as.factor(), w których argumentem musi być cała kolumna. 

**!** Klasę danych w R studio dla poszczewgólnych kolumn można sprawdzić poprzez wyświetlenie tabeli finkcją View() i umieszczenie kursora na nagłówku kolumny.

# Proste operacje na danych w data frames

```r
#W danych CO2_data zamień wartość w piątym wierszu kolumny Type etykietą ‘Unknown’. 
CO2_data[5,"Type"] <- "Unkonown"
CO2_data[5,]
```

```r
### Przelicz stężenia CO2 podane w kolumnie (zmiennej) conc, tak były wyrażone w jednostkach L/L.
#?CO2
# Informacje w hepl pozwoliły nam poznać jednostke conc
#conc - a numeric vector of ambient carbon dioxide concentrations (mL/L).
CO2_data[,"conc"] <- CO2_data[,"conc"]*0.001
CO2_data[,"conc"]
```

```r
#Oblicz wartość średnią dla kolumny (zmiennej) uptake.
mean((CO2_data$uptake))
#Użyj funkcji nrow() i ncol() do oblicznia całkowitej liczby ‘komórek’ w tabeli CO2_data.
nrow(CO2_data)*ncol(CO2_data)
```

```r
#W pierwszym wierszu CO2_data, ustaw wartość conc jako brakującą NA/ 
#Zwróć uwagę, że podajemy NA bez cudzysłowu, bo oznacza symbol zmiennej (choć o stałej wartości, podobnie 
#jak pi oznacza wartość π), a nie wartość znakową. 
CO2_data[1,"conc"] <- NA
CO2_data[1:2,]
```

```r
# Usuń wiersz z brakującą wartością i zapisz wynik w zmiennej CO2_data_noNA. 
CO2_data_noNA <- na.omit(CO2_data)
CO2_data_noNA[1:2,]
```

## Dodawanie nowych kolumn i wierszy

Dodawanie nowych kolumn jest proste: wystarczy mieć wektor danych, który będzie dodany jako nowa kolumna w tabeli. Jednakże, pamiętaj że:

- długość wektora musi odpowiadać liczbie wierszy w tabeli
- porządek wartości w wektorze musi być sensownie zgodny z porządkiem danych w tabeli
- nazwa wektora będzie nazwą kolumny w tabeli, więc nie może być taka, jak już istniejące kolumny

```r
#Utwórz wektor ID zawierający kolejne wartości naturalne od 1 do liczby równej liczbie wierszy w tabeli CO2_data.
ID <- c(1:nrow(CO2_data))
ID
```

```r
# Dodaj do CO2_data wektor ID jako pierwszą kolumnę i zapisz zmodyfikowaną tabelę pod nazwą CO2_data_ID.
CO2_data_ID <- cbind(ID,CO2_data)
```

```r
CO2_data_ID
```

```r
#Utwórz tabelę o nazwie CO2_add_row z kolumnami o nazwach takich jak w CO2_data, i wartościami dla jednego wiersza, dla kolejnych kolumn: 100, ‘Kr1’, ‘Krakow’, ‘unknown’, 1000, 20.
CO2_data_ID_add <- data.frame(ID=83, plant="Kr1", Type="Krakow", Treatment="unknown", conc=0.2, uptake=23.2)
CO2_data_ID_add
```

```r
colnames(CO2_data_ID_add)[2] = "Plant"
```

```r
ID <- c(1:nrow(CO2_data_noNA))
CO2_data_noNA<- cbind(ID,CO2_data_noNA)
```

```r
#Dołącz nowo utworzony rekord danych z tabeli CO2_add_row jako ostatni wiersz w tabeli CO2_data i zapisz rozszerzoną tabelę pod nazwą CO2_data_ID_add.
CO2_add_row <- rbind(CO2_data_noNA,CO2_data_ID_add)
```
